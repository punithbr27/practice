5   Use Map, Filter, Reduce and Lambda function on list using numpy



```python
import numpy as np
from functools import reduce

a = np.array(np.arange(10))
print("Original array is ",a)

print("\nMapping the array by adding every number with 2")
print(list(map(lambda i : i + 2,a)))

print("\nFiltering the array which is lesser than 5")
print(list(filter(lambda i : i<5,a)))

print("\nReducing the array by adding them all")
print(reduce(lambda a,b: a+b,a))
```

20 Create two series as hown using pd.Series() function. Series A = (10,20,30,40,50) and Series B = (40,50,60,70,80)

`Get the items not common to the both`

`Identify the smallest and largest element in A.`

`Find the sum of Series B.`

`Calculate average in Series A`

`Find the median in the given series B`

```python
import pandas as pd

a = pd.Series([10,20,30,40,50])
b = pd.Series([40,50,60,70,80])

a1 = set(a)
b1 = set(b)

print("The items which are not common in the both are ",a1.symmetric_difference(b1))

print("\nThe smallest element in A is",a.min())
print("The largest element in A is",a.max())

print("\nSum of the series B is ",b.sum())

print("\nAverage of Series A",a.mean())

print("\nMedian in the series B is ",b.median())
```

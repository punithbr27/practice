25 Univariate and Multivariate distribution plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("C:\\Users\\Admin\\Desktop\\Folders\\Datasets\\iris.csv")

# sepal_length = data['sepal_length']
# sepal_width = data['sepal_width']
# petal_length = data['petal_length']
# petal_width = data['petal_width']
plt.hist(data['sepal_length'],bins=8,color='grey',edgecolor='k',alpha=0.4)
plt.title("Univariate distribution plot")
plt.show()

sns.boxplot(data[['petal_length','petal_width','sepal_length','sepal_width']])
plt.title("Multivariate distribution plot")
plt.show()

plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
sns.boxplot(x= data['species'],y=data['petal_length'])
plt.subplot(2,2,2)
sns.boxplot(data=data,x='species',y='petal_width')
plt.subplot(2,2,3)
sns.boxplot(data=data,x='species',y='sepal_length')
plt.subplot(2,2,4)
sns.boxplot(data=data,x='species',y='sepal_width')
plt.suptitle("Multivariate distribution plot")
plt.tight_layout()
plt.show()
```

27 Univariate and Multivariate Composition plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("C:\\Users\\Admin\\Desktop\\Folders\\Datasets\\iris.csv")

sl = data.groupby('species')['sepal_length'].mean()
plt.pie(sl,autopct='%0.2f%%',labels=['setosa','versicolor','virginica'])
plt.legend(loc='lower left')
plt.title("Univarite composition plot")
plt.show()

data = pd.read_csv("C:\\Users\\Admin\\Desktop\\Folders\\Datasets\\iris.csv").head(20)

sl = data.groupby('species')[['sepal_length','sepal_width','petal_length','petal_width']]

plt.figure(figsize=(8,4))
sl.plot(kind='bar',stacked=True)
plt.title("Multivariate Composition plot")
plt.tight_layout()
plt.show()

print(list(sl))
```

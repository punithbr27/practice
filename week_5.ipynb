34 Univariate outliers and  Multivariate outliers

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("C:\\Users\\Admin\\Desktop\\Folders\\Datasets\\card_transdata.csv").head(50)
print("Number of missing values in the dataset")
print("\n",data.isna().sum())

sns.boxplot(data['distance_from_home'])
plt.title("Univariate outlier detection")
plt.xlabel("Variable")
plt.ylabel("Distace of the card holders home from the bank")
plt.show()


q1 = data['distance_from_home'].quantile(0.25)
q3 = data['distance_from_home'].quantile(0.75)

IQR = q3-q1

lower_bound = q1-1.5*IQR
upper_bound = q3+1.5*IQR

outliers = data[(data['distance_from_home']<=lower_bound) | (data['distance_from_home']>=upper_bound)]
print(outliers)

sns.boxplot(data[['distance_from_home','distance_from_last_transaction','online_order']])
plt.show()
```
